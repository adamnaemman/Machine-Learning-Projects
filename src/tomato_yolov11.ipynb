## 1.0 Executive Introduction & Optimization Objectives

In this stage, we utilize State-of-the-Art (SOTA) computer vision algorithms by training a **YOLOv11 Nano Classifier (`yolov11n-cls`)**. By leveraging, our model builds upon features previously learned from millions of general-world images and refines them to distinguish the micro-textures of tomato plant diseases.

### Core Training & Validation Objectives:
1. **GPU Acceleration Activation:** Connect PyTorch directly to local NVIDIA graphics processing hardware using the CUDA toolkit to handle heavy parallel tensor operations efficiently.
2. **Execute Fine-Tuning Protocols:** Train the neural network architecture across 20 distinct epochs at a uniform $640 \times 640$ spatial input configuration using the balanced dataset splits.

## 2.0 The 6-Step Model Training & Analytics Architecture

To systematically load our processed images, train the deep vision model, and extract academic-grade performance graphics, our team implements a structured **6-step training pipeline**:

* **Step 0: Hardware Verification & Ultralytics Environment Setup** Running an immediate diagnostic environment sweep to initialize the Ultralytics API framework and verify active background package builds.
* **Step 1: Deep Learning Subsystem & Hardware Validation** Auditing our localized graphics computing hardware to ensure the PyTorch backend can safely utilize dedicated GPU CUDA cores instead of slow CPU processing.
* **Step 2: YOLOv11 Classification Training Initiation** Loading the pre-trained `yolov11n-cls.pt` weight structure, binding our configuration variables (epochs, batch constraints, input resolutions), and executing the model optimization loop.
* **Step 3: Model Evaluation and Test Dataset Predictions** Freezing our optimal saved model checkpoints (`best.pt`) and running rapid inference across our holdout test directory to extract ground-truth versus prediction vectors.
* **Step 4: Confusion Matrix Generation & Visual Reporting** Using Seaborn and Scikit-Learn to compile predictions into a $10 \times 10$ matrix heatmap, visually isolating exactly which plant diseases our model classifies perfectly and which symptoms it confuses.
* **Step 5: Classification Performance Analytics & Report Compilation** Calculating high-precision data metrics (Precision, Recall, F1-Scores) down to 4 decimal points and writing out an official, permanent text log summary sheet for our project documentation.

```python
# Step 1.0
from ultralytics import YOLO

MODEL_SIZE = 'yolo11n-cls.pt'
 
model = YOLO(MODEL_SIZE)
print(f"✅ Loaded: {MODEL_SIZE}")
print(f"   Model parameters: {sum(p.numel() for p in model.model.parameters()):,}")
 
```

```python
# Step 2.0
DATASET_YAML = "C:/Users/AHMAD ADAM HAKIMI/Downloads/Machine_Learning_Project/PlantVillage_Balanced"   # folder output preprocessing kau
EPOCHS       = 50     
BATCH_SIZE   = 16     
IMG_SIZE     = 640
LR           = 0.01   
 
print("🚀 Starting YOLOv11 training...")
print(f"   Model    : {MODEL_SIZE}")
print(f"   Epochs   : {EPOCHS}")
print(f"   Batch    : {BATCH_SIZE}")
print(f"   Img size : {IMG_SIZE}")
print(f"   Device   : {DEVICE}")
print("-"*50)
 
results = model.train(
    data    = DATASET_YAML, 
    epochs  = EPOCHS,
    imgsz   = IMG_SIZE,
    batch   = BATCH_SIZE,
    device  = DEVICE,
    lr0     = LR,              
    lrf     = 0.01,            
    momentum= 0.937,          
    weight_decay = 0.0005,     
    warmup_epochs= 3,          
    patience = 15,             
    save     = True,           
    plots    = True,          
    verbose  = True,
    project  = "tomato_yolov11",
    name     = "yolov11n",    
)
 
print("\n✅ Training complete!")
print(f"   Best model saved at: tomato_yolov11/yolov11n/weights/best.pt") 
 
```

```python
# Step 4.0
from ultralytics import YOLO
import json
 
# Load best pre-trained model 
best_model = YOLO("runs/classify/tomato_yolov11/yolov11n3/weights/best.pt")
 
print("📊 Evaluating on TEST set...")
test_results = best_model.val(
    data   = DATASET_YAML,
    split  = 'test',
    imgsz  = IMG_SIZE,
    device = DEVICE,
    plots  = True,
    project= "tomato_yolov11",
    name   = "yolov11n_test"
)
 
# Print key metrics
print("\n" + "="*50)
print("  TEST SET RESULTS — YOLOv11")
print("="*50)
print(f"  Top-1 Accuracy : {test_results.results_dict.get('metrics/accuracy_top1', 0)*100:.2f}%")
print(f"  Top-5 Accuracy : {test_results.results_dict.get('metrics/accuracy_top5', 0)*100:.2f}%")
print("="*50)
```

```python
# Step 5.0
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from pathlib import Path
import os
 
# Load best pre-trained model
best_model = YOLO("runs/classify/tomato_yolov11/yolov11n3/weights/best.pt")
 
# Collect predictions on test set
test_dir = "C:/Users/AHMAD ADAM HAKIMI/Downloads/Machine_Learning_Project/PlantVillage_Balanced/test"
y_true, y_pred = [], []
 
class_names_sorted = sorted(os.listdir(test_dir))
 
print("🔍 Running predictions on test images...")
for label_idx, cls in enumerate(class_names_sorted):
    cls_path = os.path.join(test_dir, cls)
    img_files = [f for f in os.listdir(cls_path)
                 if f.lower().endswith(('.jpg','.jpeg','.png'))]
 
    for img_file in img_files:
        img_path = os.path.join(cls_path, img_file)
        pred = best_model.predict(img_path, verbose=False, device=DEVICE)
        pred_class = pred[0].probs.top1   # index of top prediction
        y_true.append(label_idx)
        y_pred.append(pred_class)
 
# Short class labels untuk display
short_names = [c.replace('Tomato_','').replace('Tomato__','').replace('_',' ')
               for c in class_names_sorted]
 
# --- Confusion Matrix ---
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(14, 11))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=short_names, yticklabels=short_names,
            linewidths=0.5, ax=ax)
ax.set_xlabel('Predicted Label', fontsize=11)
ax.set_ylabel('True Label', fontsize=11)
ax.set_title('Confusion Matrix — YOLOv11 (Test Set)', fontsize=13, fontweight='bold')
plt.xticks(rotation=35, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.tight_layout()
plt.savefig('runs/classify/tomato_yolov11/yolov11n3/yolov11_confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved: yolov11_confusion_matrix.png")
 
# --- Classification Report ---
print("\n📋 Classification Report — YOLOv11")
print("="*70)
report = classification_report(y_true, y_pred, target_names=short_names, digits=4)
print(report)
 
# Save report as text
with open("runs/classify/tomato_yolov11/yolov11n3/yolov11_classification_report.txt", "w") as f:
    f.write("Classification Report — YOLOv11\n")
    f.write("="*70 + "\n")
    f.write(report)
print("✅ Saved: yolov11_classification_report.txt")
```
